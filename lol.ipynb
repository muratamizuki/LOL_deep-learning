```python
import pandas as pd
import requests
import configparser
import sqlite3
import time
```

```python

def init_kr_db():
    # 'kr_data.db' という名前のデータベースを作成または接続
    conn = sqlite3.connect('kr_data.db')
    c = conn.cursor()

    # summoners テーブルを作成
    c.execute('''
    CREATE TABLE IF NOT EXISTS summoners (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        account_id TEXT,
        rank TEXT
    )
    ''')

    conn.commit()
    conn.close()

if __name__ == "__main__":
    init_kr_db()
```

```python
# config.iniファイルの読み込み　api_keyの取得
config_ini = configparser.ConfigParser()
config_ini.read('config.ini', encoding='utf-8')
api_key = config_ini['DEFAULT']['key']
```

```python

# チャレンジャーのデータ(summoner_id)を取得
def get_challenger_summoners(api_key, REGION):
    url = f'https://{REGION}.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key={api_key}'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None
a = get_challenger_summoners(api_key, REGION)

```

```python
# 
def get_account_ids(api_key, REGION, summoner_ids):
    account_ids = []
    for i, summoner_id in enumerate(summoner_ids, start=1):
        url = f'https://{REGION}.api.riotgames.com/lol/summoner/v4/summoners/{summoner_id}?api_key={api_key}'
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            account_ids.append(data['accountId'])
            print(f"Summoner ID: {summoner_id} -> Account ID: {data['accountId']}")
        else:
            print(f"Error: {response.status_code} for Summoner ID: {summoner_id}")
            print(response.json())

        # リクエストの制限を考慮して待機
        if i % 100 == 0:
            time.sleep(70)
        elif i % 20 == 0:
            time.sleep(10)

    return account_ids
```

```python
summoner_data = get_account_ids(api_key, REGION, summoner_ids)
```

```python
# チャレンジャーデータをsqliteに
def info_account_id(summoner_data):
    # データベースに接続
    conn = sqlite3.connect('kr_data.db')
    c = conn.cursor()
    # データの挿入
    for summoner_data_info in summoner_data:
        c.execute('''
        INSERT OR REPLACE INTO summoners(account_id, rank)
        VALUES(?,?)
        ''', (summoner_data_info,'CHALLENGER'))

    # 変更を保存して接続を閉じる
    conn.commit()
    conn.close()
```
